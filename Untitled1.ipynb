```python
import zipfile
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report

# 1. Cari & Ekstrak file archive.zip
print("1. Meng-ekstrak dataset...")
zip_path = 'archive.zip'
if not os.path.exists(zip_path):
    zip_path = '/archive.zip'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('flowers_data')

# Menentukan lokasi folder dataset
DATASET_DIR = 'flowers_data'
if 'flowers' in os.listdir(DATASET_DIR):
    DATASET_DIR = os.path.join(DATASET_DIR, 'flowers')

CATEGORIES = ["daisy", "dandelion", "rose", "sunflower", "tulip"]

# 2. Preprocessing & Ekstraksi Fitur
print("2. Memproses gambar...")
IMG_SIZE = 64
X, y, image_paths = [], [], []

for category_idx, category in enumerate(CATEGORIES):
    folder_path = os.path.join(DATASET_DIR, category)
    if not os.path.exists(folder_path):
        continue
    for img_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_name)
        try:
            img = cv2.imread(img_path)
            if img is None:
                continue
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img_resized = cv2.resize(img_rgb, (IMG_SIZE, IMG_SIZE))
            features = img_resized.flatten() / 255.0

            X.append(features)
            y.append(category_idx)
            image_paths.append(img_path)
        except Exception:
            pass

X = np.array(X)
y = np.array(y)

# 3. Pembagian Dataset (80% Train, 20% Test) Menggunakan NIM Kamu
NIM_KAMU = 24146073

X_train, X_test, y_train, y_test, path_train, path_test = train_test_split(
    X, y, image_paths, test_size=0.2, random_state=NIM_KAMU
)

# 4. Melatih Model Multi-Layer Perceptron (MLP)
print("3. Melatih model MLP (proses ini memakan waktu 1-2 menit, tunggu ya)...")
mlp = MLPClassifier(hidden_layer_sizes=(128, 64), max_iter=300, random_state=NIM_KAMU)
mlp.fit(X_train, y_train)

# 5. Menampilkan Classification Report (Precision, Recall, F1-Score, Akurasi)
y_pred = mlp.predict(X_test)
print("\n" + "="*20 + " CLASSIFICATION REPORT " + "="*20)
print(classification_report(y_test, y_pred, target_names=CATEGORIES, digits=4))

# 6. Menampilkan 25 Gambar Hasil Prediksi (5 Gambar per Kelas)
fig, axes = plt.subplots(5, 5, figsize=(13, 13))
plt.subplots_adjust(hspace=0.6)

for i, category in enumerate(CATEGORIES):
    indices_in_class = [idx for idx, label in enumerate(y_test) if label == i]
    selected_indices = random.sample(indices_in_class, 5)

    for j, idx in enumerate(selected_indices):
        ax = axes[i, j]
        img_orig = cv2.imread(path_test[idx])
        img_orig = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)

        true_label = CATEGORIES[y_test[idx]]
        pred_label = CATEGORIES[y_pred[idx]]

        ax.imshow(img_orig)
        ax.axis('off')
        color = 'green' if true_label == pred_label else 'red'
        ax.set_title(f"True: {true_label}\nPred: {pred_label}", color=color, fontsize=9)

plt.show()
```
